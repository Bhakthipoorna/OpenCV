Face and eye detector

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

face_classifier=cv2.CascadeClassifier('Haarcascades\haarcascade_frontalface_default.xml')
eyes_classifier=cv2.CascadeClassifier('Haarcascades\haarcascade_eye.xml')

imgg1=cv2.imread('Images\me2.jpg') 
img4 = cv2.resize(imgg1, (960, 540)) #Since the image is too big, we are resizing image

gray = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY) 

faces = face_classifier.detectMultiScale(gray, 1.0485258, 7)

if faces is ():
    print("No face found")
    
for (x,y,w,h) in faces:
    cv2.rectangle(img4, (x,y), (x+w,y+h), (100,0,255), 3)
    cv2.imshow("Face Detection", img4)
    cv2.waitKey(5000)
    roi_gray=gray[y:y+h,x:x+w]
    roi_color=img4[y:y+h,x:x+w]
    eyes = eyes_classifier.detectMultiScale(roi_gray)
    for(ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2)
        cv2.imshow("Eyes Detection", img4)
        cv2.waitKey(5000)
    
cv2.destroyAllWindows()   
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

face_classifier=cv2.CascadeClassifier('Haarcascades\haarcascade_frontalface_default.xml')
eyes_classifier=cv2.CascadeClassifier('Haarcascades\haarcascade_eye.xml')


ims=cv2.imread('Images\group.jpg') 
img = cv2.resize(ims, (960, 540)) 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

faces = face_classifier.detectMultiScale(gray, 1.0485258, 7)

if faces is ():
    print("No face found")
    
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (127,0,255), 2)
    cv2.imshow("Face Detection", img)
    cv2.waitKey(3000)  
    roi_gray=gray[y:y+h,x:x+w]
    roi_color=img[y:y+h,x:x+w]
    eyes = eyes_classifier.detectMultiScale(roi_gray)
    for(ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2)
        cv2.imshow("Eyes Detection", img)
        cv2.waitKey(5000)
cv2.destroyAllWindows()   
```
